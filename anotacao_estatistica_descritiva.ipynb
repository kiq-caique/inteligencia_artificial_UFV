# VISÃO ESPACIAL DAS ESTATÍSTICAS DESCRITIVAS

**Sepala**

```python
x= dados['sepal_length']
y= dados['sepal_width']
```

```python
plt.scatter(x,y)
plt.xlabel("sepal_lengt", size= 20)
plt.ylabel("sepal_width", size= 20)
mean_x = np.mean(x)
mean_y = np.mean(y)



# Desenhar o ponto da MÉDIA dentro do espaço amostral
plt.plot(np.mean(x),np.mean(y),marker='o', color='red',linestyle=':')
# Desenhar linha vertical da média até o eixo x
plt.plot([mean_x, mean_x], [np.min(y), mean_y], linestyle=':', color='red')
# Desenhar linha horizontal da média até o eixo y
plt.plot([np.min(x), mean_x], [mean_y, mean_y], linestyle=':', color='red')
# Desenhar linha diagonal da média até os mínimos do espaço amostral
plt.plot([np.min(x), mean_x], [np.min(y), mean_y], linestyle=':', color='red')


# Desenhar o ponto da MEDIANA dentro do espaço amostral
plt.plot(np.median(x),np.median(y),marker='o', color='purple')
plt.plot([np.min(x), np.median(x)], [np.median(y), np.median(y)], linestyle=':', color='purple')
plt.plot([np.median(x), np.median(x)], [np.median(y), np.min(y)], linestyle=':', color='purple')


# Calcular a MODA
modal_sl = dados['sepal_length'].mode().iloc[0]
modal_sw = dados['sepal_width'].mode().iloc[0]

mode_x = modal_sl
mode_y = modal_sw

plt.plot(mode_x, mode_y, marker='o', color='green')
plt.plot(mode_x, mode_y, marker='o', color='green')
plt.plot([np.min(x), mode_x], [mode_y, mode_y], linestyle=':', color='green')
plt.plot([mode_x, mode_x], [np.min(y), mode_y], linestyle=':', color='green')

plt.show()
```

# VARIÂNCIA E DESVIO PADRÃO

```python
# Variância Pandas
dados.var()
```

```python
# Variância Numpy
np.var(dados)
```

A diferença nos valores de variância calculados por dados.var() e np.var(dados) ocorre porque eles utilizam, por padrão, métodos ligeiramente diferentes para calcular a variância:

dados.var() do Pandas: Calcula a variância amostral, usando o divisor N-1, onde N é o número de observações. Isso é conhecido como variância "correta" ou "não-viesada", pois ajusta a amostra para melhor estimar a variância da população.

np.var(dados) do NumPy: Calcula a variância populacional por padrão, usando o divisor N, que é o número de observações.

Para alinhar os resultados, você pode:

```python
import numpy as np

#Fazer com que np.var() use o divisor N-1 (variância amostral):
var_amostral = np.var(dados, ddof=1)

#Calcular a variância populacional com o Pandas, se for isso que você deseja:
var_populacional = dados.var(ddof=0)

var_amostral, var_populacional
```

**DESVIO PADRÃO**

A diferença entre os valores dos desvios padrão calculados por np.std(dados) e dados.std() ocorre pelo mesmo motivo que a diferença nas variâncias: o divisor utilizado no cálculo.

dados.std() do Pandas: Calcula o desvio padrão amostral, usando o divisor N-1, onde N é o número de observações. Esse ajuste é para a variância amostral, o que torna o desvio padrão "não-viesado" para amostras.

np.std(dados) do NumPy: Calcula o desvio padrão populacional por padrão, usando o divisor N.

```python
np.std(dados), dados.std()
```

ddof=1 ajusta o cálculo para o desvio padrão amostral.

ddof=0 calcula o desvio padrão populacional.

```python
import numpy as np

# Fazer com que np.std() use o divisor N-1 (desvio padrão amostral):
desvio_padrao_amostral = np.std(dados, ddof=1)
print(desvio_padrao_amostral)

# Calcular o desvio padrão populacional com o Pandas:
desvio_padrao_populacional = dados.std(ddof=0)
print(desvio_padrao_populacional)
```
