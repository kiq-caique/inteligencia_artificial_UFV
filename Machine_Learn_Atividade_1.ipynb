```python
path = '/content/melb_data.csv'

df_housing = pd.read_csv(path)
print('Dataset carregado!')

# Removendo colunas qualitativas que não serão utilizadas no exercício
df_housing = df_housing.drop([ 'BuildingArea', 'YearBuilt','Suburb', 'Address', 'Type', 'Method', 'SellerG', 'Date', 'CouncilArea', 'Regionname'], axis=1)
df_housing.head()
```

```python
#removendo NAs relativos às colunas restantes
df_housing = df_housing.dropna()

# Conferindo a matrix de correlação atual
corr_matrix = df_housing.corr(numeric_only=True)
corr_matrix
```

```python
# Correlação entre Rooms e Distance
df_housing["rooms_distance"] = df_housing["Rooms"]/df_housing["Distance"]

#correlações
corr_matrix=df_housing.corr()
corr_matrix['Price'].sort_values(ascending=False)
```
